```python
from google.colab.patches import cv2_imshow
import cv2
image = cv2.imread("/content/coichitay.jpg")
cv2_imshow(image) #to view the palm in python
cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 120,apertureSize = 3)
cv2_imshow(edges)
cv2.waitKey(0)
```

```python
edges = cv2.bitwise_not(edges)
cv2.imwrite("/content/coichitay.jpg", edges)
palmlines = cv2.imread("//content/coichitay.jpg")
img = cv2.addWeighted(palmlines, 0.3, image, 0.7, 0)
cv2_imshow(img) #to view the palm in python
```
